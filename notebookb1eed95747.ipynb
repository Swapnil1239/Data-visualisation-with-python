```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df_sa=pd.read_csv("../input/police-department-incidents-previous-year-2016/Police_Department_Incidents_Previous_Year_2016.csv")
df_sa.head()
```

```python
df_sa.rename(columns={'PdDistrict':'Neighborhood'},inplace=True)
df_sa=df_sa.groupby('Neighborhood').size().reset_index(name='Count')
df_sa
```

```python

!wget --quiet https://cocl.us/sanfran_geojson -O world_countries.json

sanfran_geo = r'world_countries.json' 
```

```python
sa_map=folium.Map(location=[37.77,-122.42],zoom_start=12)
sa_map.choropleth(
    geo_data=sanfran_geo,
    data=df_sa,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)


sa_map
```
